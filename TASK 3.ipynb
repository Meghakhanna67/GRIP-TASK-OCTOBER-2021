```python
#importing all the required libraries
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
%matplotlib inline
```

```python
#importing the dataset
iris=pd.read_csv("Iris.csv")
```

```python
iris.head()
```

```python
iris['Species']=iris["Species"].str.split('-',expand=True)[1]
```

```python
iris.drop('Id',axis=1,inplace=True)
iris.head()
```

```python
iris.shape
```

```python
iris.info()
```

```python
iris.describe()
```

```python
#DATA VISUALIZATION USING HEATMAP
iris.corr()
sns.heatmap(iris.corr(),annot=True,cmap='coolwarm')

```

Train Test Split

Let's split up the data into a training set and a test set!

```python
from sklearn.model_selection import train_test_split
```

```python
X=iris.iloc[:,:4]
y=iris["Species"]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=101)
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

PREDICTION AND EVALUATION

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

TREE VISUALIZATION

```python
from sklearn import tree
```

```python
feature_names=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
class_names=['setosa','versicolor','virginica']
```

```python
fig, ax =plt.subplots(nrows=1,ncols=1,figsize=(7,7),dpi=300)
tree.plot_tree(dtree,feature_names=feature_names,class_names=class_names,filled=True)
fig.savefig('DecisionTree_Iris.png')
```
